```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


warnings.filterwarnings("ignore")
df = pd.read_csv("../data/merged.csv")
```

## Univariate Charts

```python
# Chart - 1 visualization code
facecolor = 'lightskyblue'
fig, axes = plt.subplots(ncols=2, nrows=1, facecolor=facecolor, figsize=(20,5))
sns.distplot(df['Sales'], ax=axes[0]) # Without log-transformation
axes[0].set_title('Distribution without log transformation')
sns.distplot(np.log2(df['Sales']+1), ax=axes[1]) # After log-transformation
axes[1].set_title('Distribution after log transformation')
plt.show()
```

## 📊 Sales Distribution Analysis

- Sales distribution is highly right-skewed with extreme outliers
- Log transformation makes the distribution more symmetric and reduces impact of outliers
- Log-transformed Sales is more suitable for modeling

## Bi-variate

```python
sns.scatterplot(x="Customers", y="Sales", data=df, alpha=0.3)
plt.title("Customer vs Sales")
plt.show()
```

## 👥 Customers vs Sales

- Strong positive relationship between Customers and Sales
- Higher customer count directly leads to higher sales
- Customers is a key feature for prediction

```python
sns.boxplot(x="Promo", y="Sales", data=df)
plt.title("Promo vs Sales")
plt.show()
```

## 📢 Promotion Impact on Sales

- Sales are significantly higher when Promo is active
- Non-promo days show lower and more consistent sales
- Promotions have a strong positive impact on sales → important feature for modeling

```python
sns.barplot(x="DayOfWeek", y="Sales", data=df)
plt.title("Day of Week vs Sales")
plt.show()
```

## 📅 Day of Week vs Sales

- Sales vary significantly across different days of the week
- Higher sales observed at the beginning and end of the week
- Clear weekly seasonality → DayOfWeek is an important feature

```python
sns.boxplot(x="StateHoliday", y="Sales", data=df)
plt.title("StateHoliday vs Sales")
plt.show()
```

## 🎉 State Holiday vs Sales

- Sales vary across different state holidays
- Certain holidays show higher sales compared to regular days
- Holidays have noticeable impact on sales → important feature for modeling

```python
sns.scatterplot(x="CompetitionDistance", y="Sales", data=df, alpha=0.3)
plt.title("Competition Distance vs Sales")
plt.show()
```

## 📍 Competition Distance vs Sales

- Weak relationship between competition distance and sales
- No clear linear pattern observed
- CompetitionDistance alone may not strongly influence sales

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(18,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## 🔗 Correlation Analysis

- Strong positive correlation between Customers and Sales
- Promo shows moderate positive correlation with Sales
- CompetitionDistance has weak correlation with Sales
